```python
import pandas as pd
import numpy as np
import datetime
from datetime import datetime
```

```python
df=pd.read_csv('/Users/zabihbuda/Desktop/York Universiy files/Data Science /schulich_data_science_1/module_5_feature_engineering_inclass.csv')
df.head()
```

```python

df=df.drop(['user_id'], axis=1)
df.head()
```

```python
print('max: ', df['total_spent'].max())
print('min: ', df['total_spent'].min())
```

```python
df['signup_date']=pd.to_datetime(df['signup_date'])
df['last_purchase_date']=pd.to_datetime(df['last_purchase_date'])
df['year']=df['last_purchase_date'].dt.year
df['month']=df['last_purchase_date'].dt.month
df['dom']=df['last_purchase_date'].dt.daysinmonth
df['weekday']=df['last_purchase_date'].dt.dayofweek
df['age_cat']=df['age'].apply(lambda x: 'teenager' if 0< x <18 else 'adult' if 19 < x < 40  else 'mid-age' if 41 < x < 60 else 'old' )
df['unit_price']=df['total_spent']/df['total_purchases']
df['spending_cat']=df['total_spent'].apply(lambda x: 'sporadic' if 0 < x < 200 else 'casual' if 201< x < 500 else 'regular' if 501 < x < 1000 else 'loyal')

df.head()
```

```python
df_dummies=pd.get_dummies(df[['favorite_product_category', 'gender', 'age_cat', 'spending_cat']])
df_dummies.head()
```
