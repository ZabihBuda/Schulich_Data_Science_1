```python
df_event=df['event_type'].value_counts().to_frame()
df_event=df_event.reset_index()
df_event.columns=['event_type', 'counts']

df_event.head()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df_event.event_type, y=df_event.counts, palette='Blues')
plt.title('Events Count')
plt.xlabel('Events')
plt.ylabel('Counts')
plt.show()
```

```python
df_category=df.groupby(['category_id'])['price'].value_counts().reset_index().head(200)
df_category.head()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(x='price',y='count', data=df_category, hue='category_id', style='category_id')
plt.show()
```
