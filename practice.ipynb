```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/Users/zabihbuda/Desktop/York Universiy files/Data Science /schulich_data_science_1/experiment_dataset.csv')
df.head()
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.isnull().sum()
```

```python
df['age_cat']=df['Age'].apply(lambda x: 'old' if 60<x<80 else 'mid-age' if 40<x<60 else 'adult' if 20<x<40 else 'teenage')
df.head()
```

```python
fig, axs=plt.subplots(1, 2, figsize=(25, 8))

sns.lineplot(ax=axs[0], x='Device', y='Time Spent', data=df)
sns.lineplot(ax=axs[1], x='Device', y='CTR', data=df)
```

```python
df.groupby(['Device'])[['Time Spent', 'CTR']].mean()
```

```python
fig, axs=plt.subplots(1, 2, figsize=(25, 8))
sns.lineplot(ax=axs[0], x='Location', y='Time Spent', data=df)
sns.lineplot(ax=axs[1], x='Location', y='CTR', data=df)
df.groupby(['Location'])[['Time Spent', 'CTR']].mean()
```

```python
fig, axs=plt.subplots(1, 2, figsize=(25, 8))
sns.lineplot(ax=axs[0], x='Variant', y='Time Spent', data=df)
sns.lineplot(ax=axs[1], x='Variant', y='CTR', data=df)
df.groupby(['Variant'])[['Time Spent', 'CTR']].mean()
```
